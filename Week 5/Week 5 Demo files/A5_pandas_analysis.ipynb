```python
import pandas as pd

# Load the MP1 dataset so I can inspect app review patterns.
# Use a local path so this notebook runs correctly from inside this folder.
df = pd.read_csv('app_reviews_demo.csv')

# Quick first look: what does one row look like and what columns are available?
# Meaning: this preview helps confirm the dataset loaded correctly before analysis.
df.head()
```

```python
# Question 1: What are the column types and how complete is each column?
# Meaning: this tells me whether each field is numeric/text and where missing data may affect conclusions.
df.info()
```

```python
# Question 2: Which star ratings are most common across all reviews?
# Meaning: this distribution shows overall sentiment and whether the dataset skews positive or negative.
rating_counts = df['rating'].value_counts().sort_index()
rating_counts
```

```python
# Question 3: How many low-rated reviews (1-2 stars) are in this dataset?
# Meaning: this isolates likely pain points and estimates how large the dissatisfied user segment is.
df_bad = df[df['rating'] < 3]
bad_pct = len(df_bad) / len(df) * 100
print(f"Low-rated reviews: {len(df_bad)} / {len(df)} ({bad_pct:.1f}%)")
df_bad.head()
```

```python
# Question 4: Which apps have the highest and lowest average ratings?
# Meaning: grouping by app shows relative product satisfaction and where improvement may be needed.
app_avg_rating = df.groupby('app')['rating'].mean().sort_values()
app_avg_rating
```

```python
# Question 5: Where are missing values concentrated?
# Meaning: missingness tells me which fields are less reliable for downstream analysis.
missing_counts = df.isnull().sum()
missing_counts
```
