```python
import requests
import pandas as pd

# 英超所有球队
url = "https://www.thesportsdb.com/api/v1/json/3/search_all_teams.php?l=English%20Premier%20League"
response = requests.get(url)
data = response.json()

# 看一眼结构
print(type(data))
print("Top-level keys:", list(data.keys()))
print(f"球队数: {len(data['teams'])}")
print(f"第一支球队的字段:\n{list(data['teams'][0].keys())[:15]}")
```
